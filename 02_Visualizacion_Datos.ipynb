## 2.1 Gráfico de línea con Matplotlib


```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Función seno")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```

Este gráfico muestra una función continua (`sin(x)`) usando `plt.plot`,
incluyendo título, etiquetas de ejes y rejilla.


## 2.2 Gráfico de barras con Matplotlib


```python
categorias = ["A", "B", "C", "D"]
valores = [23, 17, 35, 29]

plt.bar(categorias, valores)
plt.title("Ventas por categoría")
plt.xlabel("Categoría")
plt.ylabel("Unidades")
plt.show()
```

## 2.3 Histograma con Matplotlib


```python
# Datos aleatorios con distribución normal
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30)
plt.title("Histograma de datos normales")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

## 3.1 Análisis visual del dataset Titanic con Seaborn


```python
import seaborn as sns

titanic = sns.load_dataset("titanic")
titanic.head()
```

### 3.1.1 Supervivencia por sexo


```python
sns.countplot(data=titanic, x="sex", hue="survived")
plt.title("Supervivencia por sexo")
plt.xlabel("Sexo")
plt.ylabel("Cantidad")
plt.show()
```

### 3.1.2 Distribución de edad por clase


```python
sns.boxplot(data=titanic, x="class", y="age")
plt.title("Distribución de edad por clase")
plt.xlabel("Clase")
plt.ylabel("Edad")
plt.show()
```

### 3.1.3 Matriz de correlación


```python
corr = titanic.select_dtypes(include=["int64", "float64"]).corr()

sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Matriz de correlación - Titanic")
plt.show()
```

## 4. Gráficos interactivos con Plotly


```python
import plotly.express as px

iris = sns.load_dataset("iris")

fig = px.scatter(
    iris,
    x="sepal_length",
    y="sepal_width",
    color="species",
    title="Iris - Sepal length vs width (interactivo)"
)
fig.show()
```
